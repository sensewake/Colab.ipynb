```python
from nltk.corpus import brown as cb
from nltk.corpus import gutenberg as gb
```

```python
# 용량이 큰 관계로 다운로드를 한다.
```

```python
nltk.download('brown')
nltk.download('gutenberg')
```

```python
cb.fileids()
# 여러가지 파일 형태로 되어 있다는 것을 확인해 볼 수 있다.
```

```python
gb.fileids()
# gb에도 여러가지 파일 형태로 되어 있다.
```

```python
cb.categories()
# 카테고리 확인
```

```python
# 총 갯수 확인
brown_words = cb.words()
print('brown corpus의 총 단어 갯수 : ', len(brown_words))
```

```python
# gutenberg 의 총 개수 확인
guten_words = gb.words()
print('Gutenburg corpus의 총 단어 갯수 : ', len(guten_words))
```

```python
# brown corpus categorial 별로 단어를 확인해 보기
#cb.words?
# 특정 Categori에 있는 단어 출력
print(cb.words(categories='adventure'))

#특정 fileids에 있는 단어 출력
print(cb.words(fileids='ca01'))
```

```python
# gutenburg corpus file 별로 단어를 확인해 보기
gb.words()
```

```python
gb.words(fileids='austen-emma.txt')
```

```python
gb.fileids()
```

```python
# text 전체를 가져옴
raw_text = cb.raw( fileids='ca01')
raw_text
```

```python
#카테고리별로 텍스트 가져오기
raw_text = cb.raw(categories='adventure')
raw_text
```
